# For sorting 

```python
def sort(ranked_sol):
    for i in range(len(ranked_sol)):

        for j in range(0, len(ranked_sol) - i - 1):

          # compare two adjacent elements
          # change > to < to sort in descending order
          if ranked_sol[j][0] < ranked_sol[j + 1][0]:

            # swapping elements if elements
            # are not in the intended order
            temp = ranked_sol[j]
            ranked_sol[j] = ranked_sol[j+1]
            ranked_sol[j+1] = temp
    return ranked_sol

        
```

# Mutation


```python
def mutation(gen):
    random_chrom = random.randint(0,39)
    rand_index = random.randint(0,39)
    
    cc = gen[random_chrom]
    #print(gen)
    temp = cc[rand_index]
    #print(type(temp))
    #print(temp)
    if temp == 1:
        cc[rand_index] = 0
    else:
        cc[rand_index] = 1
        
    gen[random_chrom] = cc
        
    return gen

Chromosomes = create_chromosome()
mutation(Chromosomes)
```
